```python
import matplotlib.pyplot as plt
import numpy as np

# pagalbinė komanda, leidžianti braižyti grafikus Jupyter Notebook programoje
%matplotlib inline
```

<h2 style="color:seagreen">Plotting a Function</h2>

1. Create a function $f(x) = 17 \times x^3 - 25 \times x^2 + 28 \times x - 100$ 
2. Create a list x = [-10, -5, -2, 0, 2, 5, 10]
3. Calculate the value $f(x)$ at every point of $x$ 
4. Create a plot of $f(x)$

```python
# Funkcija

def f(xpoint):
    return 17 * xpoint**3 - 25 * xpoint**2 + 28 * xpoint - 100
```

```python
# vektorius x

vx = [-10, -5, -2, 0, 2, 5, 10]
```

```python
# Funkcijos reikšmių skaičiavimas x taškuose

vy = [] 

for point in vx:
    # funkcijos vertės skaičiavimas taške xpoint
    ypoint = f(point)
    # skaičiavimo rezultato išsaugojimas vektoriuje y
    vy.append(ypoint)
    
print(vy)
```

```python
# Grafikas y = f(x) 

plt.figure(figsize=(14, 8))

plt.plot(vx, vy, linewidth=3, color="steelblue")
plt.title("\nFukcijos grafikas: Y = f(X)\n", fontsize=22, weight="bold")
plt.xlabel("X", fontsize=16, weight="bold")
plt.ylabel("Y", fontsize=16, weight="bold")


# X ir Y ašys
plt.axvline(x=0, linestyle="solid", color="red", linewidth=1)  
plt.axhline(y=0, linestyle="solid", color="red", linewidth=1)

# Tinklelis
plt.grid()

# Ašių skalės
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Pavaizduokime X vektoriaus taškus
plt.scatter(vx, vy, s=100, color="lime")

plt.show()
```

**Problema:** funkcija $f(x) = 17 \times x^3 - 25 \times x^2 + 28 \times x - 100$ yra tolydi ir nepertraukiama kreivė, o mūsų grafike yra pavaizduota laužtinė. :( 

**Klausimas:** Ar mes galime pagerinti mūsų grafiką?

**Atsakymas:** Taip! Mums reikia daugiau taškų. Kuo daugiau taškų nauduosime, tuo grafikas bus tikslesnis.


<h2 style="color:green">Pagerintas grafikas</h2>

Mes padalinsime intervalą [-10, 10] į 100 dalių, ir nubraižysime grafiką iš naujo.


```python
# Vektorius X

vx1 = np.linspace(-10, 10, 101)

# Vektorius Y

vy1 = [] 

for point in vx1:
    # funkcijos vertės skaičiavimas taške xpoint
    ypoint = f(point)
    # skaičiavimo rezultato išsaugojimas vektoriuje y
    vy1.append(ypoint)

    
```

```python
# Grafikas y = f(x) 

plt.figure(figsize=(14, 8))

plt.plot(vx1, vy1, linewidth=3, color="steelblue")
plt.title("\nFukcijos grafikas: Y = f(X)\n", fontsize=22, weight="bold")
plt.xlabel("X", fontsize=16, weight="bold")
plt.ylabel("Y", fontsize=16, weight="bold")


# X ir Y ašys
plt.axvline(x=0, linestyle="solid", color="orangered", linewidth=1)  
plt.axhline(y=0, linestyle="solid", color="orangered", linewidth=1)

# Tinklelis
plt.grid()

# Ašių skalės
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.show()
```

<h2 style="color:green">Nepriklausomi kintamieji</h2>

```python
# Taškų skaičius

N = 100

# Vektorius X

vx2 = np.linspace(-10, 10, N)

# Vektorius Y

vy2 = np.random.randn(100)
```

```python
# Grafikas y = f(x) 

plt.figure(figsize=(14, 8))

plt.scatter(vx2, vy2, s=50, linewidth=3, color="steelblue", alpha=0.5)
plt.title("\nNepriklausomi kintamieji X ir Y\n", fontsize=22, weight="bold")
plt.xlabel("X", fontsize=16, weight="bold")
plt.ylabel("Y", fontsize=16, weight="bold")


# X ir Y ašys
plt.axvline(x=0, linestyle="solid", color="orangered", linewidth=1)  
plt.axhline(y=0, linestyle="solid", color="orangered", linewidth=1)

# Tinklelis
plt.grid()

# Ašių skalės
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.show()
```

<h2 style="color:green">Priklausomi kintamieji: linijinė priklausomybė</h2>

```python
# Taškų skaičius

N = 100

# Vektorius X

vx3 = np.linspace(-10, 10, N)

# Vektorius Y

vy3 = [2*x + 8 + 3*np.random.randn(1) for x in vx2]

# Regresijos linija

a, b = np.polyfit(vx3, vy3, 1)

vfit = a * vx3 + b
```

```python
# Grafikas y = f(x) 

plt.figure(figsize=(14, 8))

plt.scatter(vx3, vy3, s=50, color="orangered")
plt.title("\nLinijinė priklausomybė\n", fontsize=22, weight="bold")
plt.xlabel("X", fontsize=16, weight="bold")
plt.ylabel("Y", fontsize=16, weight="bold")


# X ir Y ašys
plt.axvline(x=0, linestyle="solid", color="orangered", linewidth=1)  
plt.axhline(y=0, linestyle="solid", color="orangered", linewidth=1)

# Tinklelis
plt.grid()

# Ašių skalės
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Regresijos linija
plt.plot(vx3, vfit, color="steelblue", linewidth=3, alpha=0.35)

plt.show()
```
